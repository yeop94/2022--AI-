### 데이터에 라벨이 다 붙어있으므로 지도학습임. 
### single perceptron ---> input layer - hidden layer - output layer
### dot product(두 점의 곱, 내적)


```python

#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@

print("[PerceptronClassifier-03-Compute:#1] (Perceptron Classifier) Read Data >> ")
input()
      # [x1Val, x2Val, Label]
Dat = [ [-8,1,'B'],  [-6,-1,'B'], [-5,-3,'B'], [-5,3,'B'], [-3,-2,'B'],
        [-2,-1,'B'], [-2,2,'B'],  [0,-3,'B'],  [2,-2,'B'], [3,7,'A'],
        [4,-4,'B'],  [4,9,'A'],   [6,4,'A'],   [6,8,'A'],  [7,-1,'A'],
        [8,2,'A'],   [9,4,'A'],   [10,-5,'A'], [12,1,'A'], [13,3,'A']
      ]



Wa,Wb,Wc = input(("input weights(Wa,Wb,Wc)(e.g.,0, 1, 0.5) >>"    )).split()
Wa = float(Wa)
Wb = float(Wb)
Wc = float(Wc)

#Wa = 0
#Wb = 1
#Wc = 0.5


print("Weights >>","Wa=",Wa,"Wb=",Wb,"Wc=",Wc)
    # slope -->  x2 = -w1/w3*x1 - w1/w3

slope = -(Wb/Wc)          # -2
intercept = -(Wa/Wc)      # 0
   #test; (-8,1)
#x1 = -8
#x2 = slope*x1 + intercept   # 16
#print("slope =",round(slope,1),",intercept = ",intercept)
input()

L = 0.2      # Learning rate

while(1):

  #print("Now, Classifying...")
  nErr = 0     # nbr of error
  iter = 0     # iter count

  for D in Dat:
    iter += 1
    x1 = D[0]   # = x1Val
    x2Val = D[1]
    x2 = slope*x1+intercept      #e.g..; x1 = -8, x2 = 16 (x2Val = 1)
    #print("x1 = ",x1,",x2Val = ",x2Val,",x2 = ",x2)
    if (x2Val > x2):
      group = 'A'
    else :
      group = 'B'
  
    #print("group = ",group)

    if (group != D[2]):      #error occurs
      #print("Current Data =",D,"computed group = ",group,"real group =",D[2])
      nErr += 1
      if (nErr == 1):  #first error
        firstE = iter - 1



    if (group == 'A'):
        dVal = -1
    else:
        dVal = 1

  #print("dVal = ",dVal)

  if nErr == 0 :
    break
    
  x1Dat = Dat[firstE][0]  # (2,-2)   --> x1 = 2
  x2Dat = Dat[firstE][1]  # (2,-2)   --> x2 = -2
  Wa = Wa + L*dVal*1
  Wb = Wb + L*dVal*x1Dat
  Wc = Wc + L*dVal*x2Dat

  slope = -(Wb/Wc)          
  intercept = -(Wa/Wc)  
  #print("Weights >>","Wa=",Wa,"Wb=",Wb,"Wc=",Wc)
  #print("slope =",round(slope,2),",intercept = ",round(intercept,2))

  x2 = slope*x1Dat + intercept
  #print("x2 = ",round(x2,2))
  #print("End.while: nErr = ",nErr)
  
print("================== Result ===============")
print("slope = ",round(slope,2),"intercept = ",round(intercept,2))

 # Classifier
while (1):
  print("\n==========  LINEAR CLASSIFIER  ==========")
  print("slope=",round(slope,2)," intercept=",round(intercept,2))
  input()
  print("Input x and y  (e.g.) 7 7  >> ")
  x, y = input().split()
  x, y = int(x), int(y)
  print("Your Input; x=",x," y=",y)
  input()
  x2 = slope * x + intercept
  if (y > x2):
    print('Classification Result= Group A')
  else:
    print('Classification Result= Group B')
  #if
  print("\nPress 1 to Continue, 0 to Quit >> ")
  ans = int(input())
  if (ans==0):
     break
  #if.0
#while.1




print("###########  END OF PROGRAM   ############")

#//
```
