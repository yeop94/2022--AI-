```python
#@@@@@@@@@@@@@@@@@@@@@@@@@@@

print("[LR-04-SSE.py] SSE >> ")
# compute sum of squares of errors

import numpy as np

#====================  SHOW_2Dim()  =====================
def SHOW_2Dim(List):  
    for L in List:        #e.g., L=[-13.6470,-22.8235]
         out=[]
         for D in L:      #e.g., [D=-13.6470, D=-22.8235]
             out.append(round(D,3))  #e.g., [-13.647,-22.824]
         #for.D
         print(out)
    #for.L
#show_2dim()

def SSE(Dat,slope):
# sse = SSE(DatN,a)
  Num = len(Dat)
  sse=0
  for i in range(0,Num):
    # error for Dat[i]
    datX = Dat[i][0]
    datY = Dat[i][1]
    error = (slope*datX-datY)**2 
    sse += error
  #for.i
  return sse
#def.sse()

##################  Main  ########################

        # Height, Weight Data; 17
Dat = [ [150,42], [153,51], [155,62], [157,50], [157,76],
        [159,45], [161,72], [162,56], [164,65], [166,50],
        [166,72], [168,75], [169,82], [170,63], [172,73],
        [176,88], [177,80]
      ]  
                     # 02-NormalizeData
Num = len(Dat)

    # (Xa, Ya); average pt 
Xsum = 0
Ysum = 0

for j in range(0,Num):   # j=0,1,2,...,Num-1
    Xsum += Dat[j][0]
    Ysum += Dat[j][1]
#for.j

Xavg = Xsum / Num    # 163.5
Yavg = Ysum / Num    # 64.8
print("Xavg=",round(Xavg,1),"  Yavg=",round(Yavg,1))
input()

DatN = np.zeros((17,2))   # init.DatN to zeros

for j in range(0,Num):
    DatN[j][0] = Dat[j][0]-Xavg
    DatN[j][1] = Dat[j][1]-Yavg
#for.j
#print("DatN[]=",DatN)
print("DatN[]=")
SHOW_2Dim(DatN)
input()
                       # 03-slope
 # four areas; area1(+,+), area2(-,+), area3(-,-), area4(+,-)
 #             Ans; 7,2,6,2
area1=0
area2=0
area3=0
area4=0
for i in range(0,Num):
  #print("for.i=",i," DatN[i]=",DatN[i])
  if (DatN[i][0]>=0 and DatN[i][1]>=0):
     area1 += 1
  elif (DatN[i][0]<0 and DatN[i][1]>=0):
     area2 += 1
  elif (DatN[i][0]<0 and DatN[i][1]<0):
     area3 += 1
  else:
     area4 += 1
  #if
  #print("Number of data pts of Areas; [",area1,area2,area3,area4,"]")  # 7,2,6,2
  #input()
#for.i
              # Compute slope a; a=1 or a=-1
if (area1+area3) > (area2+area4) :      # 13 > 4  ->  a=1
   a=1
else:
   a=-1
#if
print("slope: a=",a)        # Ans; a=1
input()

    #04-SSE 
  # error for DatN[0]
datX = DatN[0][0]
datY = DatN[0][1]
error = (a*datX-datY)**2 
print("SSE for DatN[0]=",round(error,1))
input()
#0th row; (datX,datY)=(-13.6, -22.8)
#error = (a*(-13.6)-(-22.8))**2 -> (-13.6+22.8)**2  ->  9.2*9.2 = 84.2 

  # call SSE
sse = SSE(DatN,a)
print("Result; sse=",round(sse,1))     # 1570.5
input()



#05-Change slope
  # change slope both ways; to increase & decrease
delta = 0.01
if (SSE(DatN,(a+delta)) < sse) :
  delta = 0.01

else :
  delta = -0.01

print(";; delta = ", delta)    # delta = 0.01


#a1 = a + delta
#a2 = a - delta              # 여기서 a + 델타인지 a - 델타인지 의미없고 델타값만 마이너스로 바꿔주면됨.
#sse1 = SSE(DatN,a1)
#sse2 = SSE(DatN,a2)
#print("[05] Change slope; a1=",a1," sse1=",round(sse1,1))
#print("      Change slope; a2=",a2," sse2=",round(sse2,1))
#input()

 # to find min SSE
sseOld = SSE(DatN,a) 
while (1):
  a += delta
  sse = SSE(DatN,a)
  #print("slope; a =",a," sse=",round(sse,1))
  #input()
  if (sse > sseOld) :   # terminal condition
    min_sse = sseOld
    a -= delta
    break # exit
  sseOld = sse
  #print(" ; sseOld =",round(sseOld,1))
  

    #while
print("Out of while: slope =",round(a,2),"min_sse = ",round(min_sse,1))
 

#############  Result; slope=1.27, min SSE= 1499.8 


# LR - 07 user inquiry
while(1):
  print("Solution; Y =",round(a,2)," X")      # Y = 1.27 X    
  height = int(input("Please enter a height(e,g., 171)(cm) >>" ))
  heightN = height - Xavg
  weightN = a*heightN
  weight = weightN + Yavg
  print("Expected Weight=",round(weight,1),"kg")   # 
  
  
  keyIN = input("Press 1 to continue, 0 to Quit >> ")
  if keyIN != '1' :
    break


#polynomial regression data fit  - https://arachnoid.com/polysolve/       --- 여러 데이터를 넣어보고 함수를 만들어준다(2차,3차,4차 등으로 degree를 올리면서 standard error를 줄일수잇다 
                                                                            # but, 너무 degree를 올려서 오차는 적어지지만 자칫잘못하면 과대적합(overfitting)될 수 있다. 그렇게되면 현재 데이터에만 너무 집중하여 앞으로 들어올 새로운 데이터에 적용이 안된다.)




print("###############  End of Program  ################")
```
