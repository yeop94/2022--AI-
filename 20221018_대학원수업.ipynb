# 그룹핑(clustering)
## KMC = K-Means Clustering
### 1. 두 그룹으로 나누는걸 하기 전에 먼저 그냥 랜덤한 수 두개(r1, r2)를 뽑아서 두 그룹을 대충 만들자
### 2. idea = 같은 그룹의 데이터들은 다른 그룹의 데이터들보다는 좀 더 가깝게 있다




# 과제 = 베스트 솔루션(가장 최적의 그룹핑)을 내기 위한 근거를 제시하시오. +알파로 코드 만들어보기



```python

import random    #randint 함수를 쓰기위해
import math      #sqrt 함수를 쓰기위해

     #KMC-01a. Height, Weight Data
Dat = [ [122, 38],    #0
        [125, 41],    #1
        [125, 40],    #2
        [130, 42],    #3
        [131, 40],    #4
        [132, 41],    #5
        [132, 45],    #6
        [133, 39],    #7
        [133, 44],    #8
        [133, 42],    #9
        [134, 40],    #10
        [135, 40],    #11
        [136, 42],    #12
        [137, 41],    #13
        [137, 44],    #14
        [139, 43],    #15
        [140, 45]  ]  #16

  #KMC-01b. r1, r2; two random numbers                   
r1 = random.randint(0,16)
r2 = random.randint(0,16)
while (r1 == r2):                       # 여기서 if함수가 아니라 while함수인 이유? 만약 r2를 새로 뽑았는데 또 같으면 if는 그냥 넘어가고 while은 또 다시 뽑게될듯?
  r2 = random.randint(0,16)
#while

print("r1=",r1," r2=",r2)
input()

  #KMC-01c. P, Q; two data pts of Dat[]
  # (Ph,Pw), (Qh,Qw)
Ph = Dat[r1][0]  
Pw = Dat[r1][1] 
Qh = Dat[r2][0] 
Qw = Dat[r2][1]
print("P=(",Ph,",",Pw,")")  
print("Q=(",Qh,",",Qw,")")


 #KMC=02. Distance. 나머지 점들과 P, Q 점 까지의 거리 구하기

# 예시 하나 들어보기
#Dh = Dat[0][0]
#Dw = Dat[0][1]
#distP = math.sqrt((Ph-Dh)**2 + (Pw-Dw)**2)      #distance btwn P and D, where P=(Ph,Pw), D = (Dh,Dw)
#distQ = math.sqrt((Qh-Dh)**2 + (Qw-Dw)**2)
#print("distP = ",round(distP,2), "distQ =",round(distQ,2))


group_P = []
group_Q = []

for i in range(0,len(Dat)):    
  Dh = Dat[i][0]
  Dw = Dat[i][1]
  distP = math.sqrt((Ph-Dh)**2 + (Pw-Dw)**2)
  distQ = math.sqrt((Qh-Dh)**2 + (Qw-Dw)**2)
  print('iter = ',i, ":(Dh,Dw) =",(Dh,Dw), ": distP = ",round(distP,2),"," "distQ =",round(distQ,2))    
  if distP < distQ:
    #print("distP is shorter than distQ")
    group_P.append(i)
  else:
    group_Q.append(i)
    #print("distQ is shorter than distP")

print("group_P = ",group_P,"group_Q = ",group_Q)

#KMC=03. New Center. 새로운 중심점 찾기

sumPh = 0
sumPw = 0
for p in group_P:
  sumPh += Dat[p][0]    # sum of heights
  sumPw += Dat[p][1]    # sum of weights

#for.p

avgPh = sumPh / len(group_P)
avgPw = sumPw / len(group_P)
print("avgPh = ",round(avgPh,1),"avgPw =",round(avgPw,1))

####################################################

sumQh = 0
sumQw = 0
for q in group_Q:
  sumQh += Dat[q][0]    # sum of heights
  sumQw += Dat[q][1]    # sum of weights

#for.q

avgQh = sumQh / len(group_Q)
avgQw = sumQw / len(group_Q)
print("avgQh = ",round(avgQh,1),"avgQw =",round(avgQw,1))



# 여기서 이제 이 avgP, avgQ의 점으로 다시 앞의 모든 과정을 거쳐 새로운 중심값을 찾아낸다.

########      KMC - 04. new centroids

iter = 0
Oldgroup_P = []
while (1):
  iter += 1
  print("iter:",iter,"------------------------------------")
  #input()
  group_P = []
  group_Q = []

  for i in range(0,len(Dat)):    
    Dh = Dat[i][0]
    Dw = Dat[i][1]
    distP = math.sqrt((avgPh-Dh)**2 + (avgPw-Dw)**2)
    distQ = math.sqrt((avgQh-Dh)**2 + (avgQw-Dw)**2)
 
    if distP < distQ:
      group_P.append(i)
    else:
      group_Q.append(i)

  # print("group_P = ",group_P,"group_Q = ",group_Q)
  
    # 04. terminal condition
  
  if group_P == Oldgroup_P:
    print("#######################################")
    print("group_P = ",group_P,"group_Q = ",group_Q)
    print("avgPh = ",round(avgPh,1),"avgPw =",round(avgPw,1))
    print("avgQh = ",round(avgQh,1),"avgQw =",round(avgQw,1))
    break


  sumPh = 0
  sumPw = 0
  for p in group_P:
    sumPh += Dat[p][0]    
    sumPw += Dat[p][1] 
  
  avgPh = sumPh / len(group_P)
  avgPw = sumPw / len(group_P)
  print("avgPh = ",round(avgPh,1),"avgPw =",round(avgPw,1))


  sumQh = 0
  sumQw = 0
  
  for q in group_Q:
    sumQh += Dat[q][0]    
    sumQw += Dat[q][1] 
 
  avgQh = sumQh / len(group_Q)
  avgQw = sumQw / len(group_Q)
  print("avgQh = ",round(avgQh,1),"avgQw =",round(avgQw,1))

  Oldgroup_P = group_P
 
 




print("###########  END OF PROGRAM   ############")
```

```python

import random  # randint()
import math     # sqrt() 

print("KMC-04-Iter >> ")
     #KMC-01a. Height, Weight Data
Dat = [ [122, 38],    #0
        [125, 41],    #1
        [125, 40],    #2
        [130, 42],    #3
        [131, 40],    #4
        [132, 41],    #5
        [132, 45],    #6
        [133, 39],    #7
        [133, 44],    #8
        [133, 42],    #9
        [134, 40],    #10
        [135, 40],    #11
        [136, 42],    #12
        [137, 41],    #13
        [137, 44],    #14
        [139, 43],    #15
        [140, 45]  ]  #16

  #KMC-01b. r1, r2; two random numbers
r1 = random.randint(0,16)
r2 = random.randint(0,16)
while (r1==r2):
    r2 = random.randint(0,16)
#while
#r1 = 8
#r2 = 7
print("r1=",r1," r2=",r2)     # 8, 7
input()

  #KMC-01c. P, Q; two data pts of Dat[]
  # (Ph,Pw), (Qh,Qw)
Ph = Dat[r1][0]          
Pw = Dat[r1][1] 
Qh = Dat[r2][0]          
Qw = Dat[r2][1]
print("P=(",Ph,",",Pw,")")      # P; [133, 44] ; 8th row
print("Q=(",Qh,",",Qw,")")     # Q; [133, 39] ; 7th row

###### Iter starts
iter = 0
OldGroupP = []
while (1):             # outermost iter
  iter += 1
  print("Iter=",iter,"  --------------------------------------------------")
  input()
      # KMC-02. Distance 
  nDat = len(Dat)    # 17
  GroupP = []
  GroupQ = []

  for i in range(0,nDat):    # i=0,1,2,...,16
    Dh = Dat[i][0]    # i-th row, 0-th col 
    Dw = Dat[i][1]   # i-th row, 0-th col
    distP = math.sqrt((Ph-Dh)**2+(Pw-Dw)**2)
    distQ = math.sqrt((Qh-Dh)**2+(Qw-Dw)**2)
    #print("Iter=",i,": (Dh,Dw)=(",Dh,",",Dw,")")
    #print("Iter=",i,": distP=",round(distP,2)," distQ=",round(distQ,2))
    #input()
           # distP < distQ
    if (distP < distQ):
        #print("distP is shorter than distQ")
        GroupP.append(i)
    else: 
        #print("distQ is shorter than distP")
        GroupQ.append(i)
    #if
    #print("  ; GroupP=",GroupP," GroupQ=",GroupQ)
    #input()
  #for.i

  print("  ; GroupP=",GroupP," GroupQ=",GroupQ)
  input()

      # 04. Terminal condition 
  if (GroupP == OldGroupP):
      break
  #if
  OldGroupP = GroupP

  sumPh = 0
  sumPw = 0
  for p in GroupP:
    sumPh += Dat[p][0]    # sum of heights
    sumPw += Dat[p][1]   # sum of weights
  #for.p
  avgPh = sumPh / len(GroupP)
  avgPw = sumPw / len(GroupP)
  print("avgPh=",round(avgPh,1)," avgPw=",round(avgPw,1))
  input()

  sumQh = 0
  sumQw = 0
  for q in GroupQ:
    sumQh += Dat[q][0]    # sum of heights
    sumQw += Dat[q][1]   # sum of weights
  #for.q
  avgQh = sumQh / len(GroupQ)
  avgQw = sumQw / len(GroupQ)
  print("avgQh=",round(avgQh,1)," avgQw=",round(avgQw,1))
  input()
             # 04. new centroids 
  Ph = avgPh
  Pw = avgPw
  Qh = avgQh
  Qw = avgQw
  #while.1


print("###########  END OF PROGRAM   ############")
```
